```python
from manim import *

config.media_width = "75%"
config.verbosity = "WARNING"
```

```python
%%manim -ql --format=gif DesignOfDynamicExperiments
# %%manim -qh DesignOfDynamicExperiments
# %%manim -qh --format=gif DesignOfDynamicExperiments

class DesignOfDynamicExperiments(Scene):
    
    def construct(self):

        

        ax = Axes(
            x_range=[0, 1, 1],
            y_range=[-1, 1, 1],
            x_length=5,
            tips=False,
            axis_config={"include_numbers": True},
        )

        ax.shift(LEFT*3.5)

        ax_labels = ax.get_axis_labels(
            Text("Time").scale(0.5), Text("Value").scale(0.5)
            )

        x1 = ValueTracker(0)
        x2 = ValueTracker(0)
        x3 = ValueTracker(0)

        def func(t,x1,x2,x3):
            return x1 + x2*(-1 + 2*t) + x3*(1 - 6*t + 6*t**2)

        func2 = lambda t: func(t,x1.get_value(),x2.get_value(),x3.get_value())

        time_traj = VMobject()
        
        def update_graph(mob):
            mob.become(ax.plot(func2,color=BLUE))

        update_graph(time_traj)
        time_traj.add_updater(update_graph)

        chart = BarChart(
            values=[x1.get_value(), x2.get_value(), x3.get_value()],
            bar_names=["Avg.", "Slope", "Curv."],
            y_range=[-1, 1, 1],
            y_length=6,
            x_length=4,
            x_axis_config={"font_size": 36},
        )

        chart.shift(RIGHT*3)

        chart_labels = chart.get_axis_labels(
            Text("Factor").scale(0.5), Text("Level").scale(0.5)
            )

        

        def chart_updater(obj):
            """An updater which continually move an object above the square.

            The first parameter (obj) is always the object that is being updated."""
            obj.change_bar_values([x1.get_value(), x2.get_value(), x3.get_value()])

        # we want the label to always reside above the square
        chart.add_updater(chart_updater)

    
        self.add(ax, ax_labels, chart, chart_labels)
        self.play(Create(time_traj))
        self.add(ax.plot(func2,color=GRAY))
        self.play(x1.animate.set_value(1))
        self.add(ax.plot(func2,color=GRAY))
        self.play(x1.animate.set_value(-1)) 
        self.add(ax.plot(func2,color=GRAY))
        self.play(x1.animate.set_value(0))   
        self.play(x2.animate.set_value(1))   
        self.add(ax.plot(func2,color=GRAY))
        self.play(x2.animate.set_value(-1))   
        self.add(ax.plot(func2,color=GRAY))
        self.play(x2.animate.set_value(0))   
        self.play(x3.animate.set_value(1))   
        self.add(ax.plot(func2,color=GRAY))
        self.play(x3.animate.set_value(-1))   
        self.add(ax.plot(func2,color=GRAY))
        self.play(x3.animate.set_value(0))   
        self.play(x1.animate.set_value(-1/3),x2.animate.set_value(-1/3),x3.animate.set_value(-1/3))   
        self.add(ax.plot(func2,color=GRAY))
        self.play(x1.animate.set_value(-1/3),x2.animate.set_value(-1/3),x3.animate.set_value(+1/3))   
        self.add(ax.plot(func2,color=GRAY))
        self.play(x1.animate.set_value(-1/3),x2.animate.set_value(+1/3),x3.animate.set_value(-1/3))   
        self.add(ax.plot(func2,color=GRAY))
        self.play(x1.animate.set_value(-1/3),x2.animate.set_value(+1/3),x3.animate.set_value(+1/3))   
        self.add(ax.plot(func2,color=GRAY))
        self.play(x1.animate.set_value(+1/3),x2.animate.set_value(-1/3),x3.animate.set_value(-1/3))   
        self.add(ax.plot(func2,color=GRAY))
        self.play(x1.animate.set_value(+1/3),x2.animate.set_value(-1/3),x3.animate.set_value(+1/3))   
        self.add(ax.plot(func2,color=GRAY))
        self.play(x1.animate.set_value(+1/3),x2.animate.set_value(+1/3),x3.animate.set_value(-1/3))   
        self.add(ax.plot(func2,color=GRAY))
        self.play(x1.animate.set_value(+1/3),x2.animate.set_value(+1/3),x3.animate.set_value(+1/3))   
        self.add(ax.plot(func2,color=GRAY))
        self.play(x1.animate.set_value(0),x2.animate.set_value(0),x3.animate.set_value(0))   
        
        
```
